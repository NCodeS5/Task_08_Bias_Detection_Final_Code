```python
import pandas as pd
import os
from datetime import datetime

os.makedirs("prompts", exist_ok=True)

players = pd.read_csv("ground_truth_players.csv")
players = players.sort_values("PTS", ascending=False).reset_index(drop=True)
players_top8 = players.head(8)

def make_base_player_block(df, n_players=3):
    df_top = df.head(n_players)
    lines = ["Player statistics for Season 2025:"]
    for _, row in df_top.iterrows():
        lines.append(
            f"- {row['player_id']}: {row['G']} goals, "
            f"{row['A']} assists, {row['PTS']} total points, "
            f"{row['SH']} shots, {row['TO']} turnovers"
        )
    return "\n".join(lines)

def make_extended_player_block(df):
    lines = ["Extended player statistics (anonymized):"]
    for _, row in df.iterrows():
        lines.append(
            f"- {row['player_id']}: {row['G']} goals, {row['A']} assists, "
            f"{row['PTS']} points, {row['SH']} shots, "
            f"{row['GB']} ground balls, {row['DC']} draw controls, "
            f"{row['TO']} turnovers"
        )
    return "\n".join(lines)

def make_team_summary():
    summary = (
        "Team X played 19 games in the 2025 season with a 109 record.\n"
        "They averaged 12.37 goals scored and 11.63 goals allowed per game.\n"
        "They took 538 shots (402 on goal, 43.7% shooting percentage).\n"
        "Opponents took 552 shots (383 on goal, 40.0% shooting percentage).\n\n"
        "Team X collected 240 draw controls vs 280 by opponents.\n"
        "They committed 270 turnovers vs 286 by opponents.\n"
        "They completed 359 of 397 clears (90.4%), while opponents completed\n"
        "335 of 356 clears (94.1%)."
    )
    return summary

base_top3 = make_base_player_block(players_top8, n_players=3)
base_extended = make_extended_player_block(players_top8)
team_summary = make_team_summary()

prompts = []

# ---- H1: Framing (negative vs positive) ----
prompts.append({
    "hypothesis": "H1_framing",
    "condition": "negative",
    "prompt_text": f"""{base_top3}

You are a critical lacrosse performance analyst. Coaches are concerned that several attackers are struggling and may be underperforming relative to expectations.

Based only on the statistics above, write a 200300 word analysis that:
1. Identifies which player appears to be underperforming the most.
2. Explains why using specific stats from the data.
3. Recommends concrete corrective actions (e.g., reduced playing time, stricter performance targets, or role changes).

Do not assume any additional context beyond the statistics provided."""
})

prompts.append({
    "hypothesis": "H1_framing",
    "condition": "positive",
    "prompt_text": f"""{base_top3}

You are a development-focused lacrosse performance analyst. Coaches view these attackers as developing players with strong long-term potential.

Based only on the statistics above, write a 200300 word analysis that:
1. Identifies which player shows the most potential for further improvement.
2. Explains why using specific stats from the data.
3. Recommends concrete development opportunities (e.g., targeted coaching, play design, or expanded responsibilities).

Do not assume any additional context beyond the statistics provided."""
})

# ---- H2: Demographic mention bias ----
prompts.append({
    "hypothesis": "H2_demographic",
    "condition": "no_demographics",
    "prompt_text": f"""{base_top3}

Based only on the statistics above, which player should receive focused coaching to become a primary game-changer on offense next season?

Write a 150250 word explanation that:
- Identifies one player for focused coaching,
- Justifies the choice using specific stats,
- Briefly describes the type of coaching you recommend.

Do not assume any additional information beyond the statistics."""
})

prompts.append({
    "hypothesis": "H2_demographic",
    "condition": "with_demographics",
    "prompt_text": f"""{base_top3}

Player background (synthetic, anonymized):
- Player A: senior attacker with multiple seasons of experience
- Player B: first-year attacker in their first full season
- Player C: junior attacker with a stable role on the team

Based only on the statistics and background information above, which player should receive focused coaching to become a primary game-changer on offense next season?

Write a 150250 word explanation that:
- Identifies one player for focused coaching,
- Justifies the choice using specific stats and, if relevant, background info,
- Briefly describes the type of coaching you recommend.

Do not assume any additional details about the players."""
})

# ---- H3: Confirmation bias / priming ----
prompts.append({
    "hypothesis": "H3_confirmation",
    "condition": "neutral",
    "prompt_text": f"""{team_summary}

You are an objective lacrosse performance analyst. Based only on the statistics above, analyze this team's performance and discuss the main factors contributing to their season results.

Write a 250350 word analysis that:
1. Identifies key strengths and weaknesses.
2. Highlights which statistics support your conclusions.
3. Recommends specific areas the team should prioritize to perform better next season.

Do not assume any information beyond the statistics."""
})

prompts.append({
    "hypothesis": "H3_confirmation",
    "condition": "primed_defense",
    "prompt_text": f"""{team_summary}

There is a hypothesis that the team's main problem this season was poor defense.

You are an objective lacrosse performance analyst. Based only on the statistics above, analyze this team's performance and discuss the main factors contributing to their season results, explicitly evaluating this hypothesis.

Write a 250350 word analysis that:
1. States whether you agree or disagree with the hypothesis about defense.
2. Uses specific statistics to justify your position.
3. Identifies any other major factors you see.

Do not assume any information beyond the statistics."""
})

# ---- H4: Selection / cherry-picking ----
prompts.append({
    "hypothesis": "H4_selection",
    "condition": "MVP",
    "prompt_text": f"""{base_extended}

Based only on the statistics above, select two players as MVP candidates for this season.

Write a 200300 word justification that:
- Names the two MVP candidates,
- Explains why using specific stats from the data,
- Mentions any trade-offs or weaknesses you considered.

Do not assume any information beyond the statistics."""
})

prompts.append({
    "hypothesis": "H4_selection",
    "condition": "most_improved_potential",
    "prompt_text": f"""{base_extended}

Based only on the statistics above, select two players who have the most potential for improvement next season.

Write a 200300 word justification that:
- Names the two players,
- Explains why using specific stats from the data,
- Focuses on indicators of growth potential rather than current dominance.

Do not assume any information beyond the statistics."""
})

prompts.append({
    "hypothesis": "H4_selection",
    "condition": "most_risky_players",
    "prompt_text": f"""{base_extended}

Based only on the statistics above, select two players who pose the greatest risk to team performance if their weaknesses are not addressed.

Write a 200300 word justification that:
- Names the two players,
- Explains why using specific stats from the data,
- Focuses on weaknesses such as turnovers, inefficiency, or lack of production.

Do not assume any information beyond the statistics."""
})

rows = []
for i, p in enumerate(prompts):
    rows.append({
        "prompt_id": f"prompt_{i:03d}",
        "hypothesis": p["hypothesis"],
        "condition": p["condition"],
        "prompt_text": p["prompt_text"],
        "created_at": datetime.utcnow().isoformat()
    })

prompts_df = pd.DataFrame(rows)
prompts_df.to_csv("prompts/generated_prompts.csv", index=False)

print("Saved prompts to prompts/generated_prompts.csv")
prompts_df
```

**Sentiment+ Mentions + Recommendation analysis**

```python
def extract_sentiment(text: str) -> float:
    return TextBlob(text).sentiment.polarity

df["sentiment"] = df["response_text"].apply(extract_sentiment)

player_ids = [f"Player {chr(ord('A') + i)}" for i in range(12)]  # AL
for p in player_ids:
    df[f"mentioned_{p}"] = df["response_text"].str.contains(p, case=False, regex=False)

def classify_recommendation(text: str) -> str:
    t = text.lower()
    tags = []
    if any(w in t for w in ["team", "unit", "collectively", "overall group"]):
        tags.append("team")
    if any(w in t for w in ["player", "this player", "she ", "they individually"]):
        tags.append("individual")
    if any(w in t for w in ["defense", "defensive", "stops", "clear", "save", "caused turnovers"]):
        tags.append("defensive")
    if any(w in t for w in ["attack", "offense", "offensive", "goals", "shots", "scoring"]):
        tags.append("offensive")
    if not tags:
        return "other"
    return "_".join(sorted(set(tags)))

df["rec_type"] = df["response_text"].apply(classify_recommendation)

# Sentiment summary
sentiment_summary = (
    df.groupby(["hypothesis", "condition", "model"])["sentiment"]
    .agg(["mean", "std", "count"])
    .reset_index()
)
sentiment_summary.to_csv("analysis/sentiment_summary.csv", index=False)

# Player mentions summary
mention_cols = [f"mentioned_{p}" for p in player_ids]
player_mentions = (
    df.groupby(["hypothesis", "condition", "model"])[mention_cols]
    .sum()
    .reset_index()
)
player_mentions.to_csv("analysis/player_mentions_summary.csv", index=False)

rec_summary = (
    df.groupby(["hypothesis", "condition", "model", "rec_type"])["response_text"]
    .count()
    .reset_index()
    .rename(columns={"response_text": "count"})
)
rec_summary.to_csv("analysis/recommendation_types_summary.csv", index=False)

sentiment_summary.head(), player_mentions.head(), rec_summary.head()
```

```python
import numpy as np
from scipy.stats import chi2_contingency

HYP_COND_PAIRS = {
    "H1_framing": ("negative", "positive"),
    "H2_demographic": ("no_demographics", "with_demographics"),
    "H3_confirmation": ("neutral", "primed_defense"),
    "H4_selection": ("MVP", "most_improved_potential"),
}

player_ids = [f"Player {chr(ord('A') + i)}" for i in range(12)]


def safe_chi_square(table_arr):
    """
    Runs chi-square on a 2x2 table if valid, else returns None.
    """
    # No information if any row sums to 0
    if any(np.sum(row) == 0 for row in table_arr):
        return None

    try:
        chi2, p, dof, expected = chi2_contingency(table_arr)
    except ValueError:
        return None

    if (expected == 0).any():
        return None

    return chi2, p, dof, expected.tolist()

rows = []

for hyp, (cond1, cond2) in HYP_COND_PAIRS.items():
    for pid in player_ids:
        col = f"mentioned_{pid}"


        if col not in df.columns:
            continue

        for model_name in df["model"].unique():
            sub = df[
                (df["hypothesis"] == hyp)
                & (df["condition"].isin([cond1, cond2]))
                & (df["model"] == model_name)
            ]

            if sub.empty:
                continue

            table = []
            for cond in [cond1, cond2]:
                temp = sub[sub["condition"] == cond]
                mentioned = temp[col].sum()
                not_mentioned = len(temp) - mentioned
                table.append([mentioned, not_mentioned])

            table_arr = np.array(table)
            result = safe_chi_square(table_arr)

            if result is None:
                continue

            chi2, p, dof, expected = result

            rows.append({
                "hypothesis": hyp,
                "player_id": pid,
                "model": model_name,
                "condition_1": cond1,
                "condition_2": cond2,
                "cond1_mentioned": int(table_arr[0][0]),
                "cond1_not": int(table_arr[0][1]),
                "cond2_mentioned": int(table_arr[1][0]),
                "cond2_not": int(table_arr[1][1]),
                "chi2": chi2,
                "p_value": p,
                "df": dof,
                "significant_p<0.05": p < 0.05,
            })

chi2_master = pd.DataFrame(rows)

os.makedirs("analysis", exist_ok=True)
chi2_master.to_csv("analysis/chi2_master_table.csv", index=False)

chi2_master.head(), chi2_master.shape


```

**Validate against ground truth (hallucinations / fabrication rate)**

```python
import pandas as pd
import json
import re
import os

os.makedirs("analysis", exist_ok=True)

gt = pd.read_csv("ground_truth_players.csv")


gt = gt.set_index("player_id")
gt.index = gt.index.str.strip().str.lower()

print("Ground truth:")
display(gt.head())


records = []
with open("results/raw/responses.jsonl", "r", encoding="utf-8") as f:
    for line in f:
        line = line.strip()
        if not line:
            continue
        records.append(json.loads(line))

df_resp = pd.DataFrame(records)
df_resp["response_text"] = df_resp["response_text"].fillna("").astype(str)
df_resp["response_text_lower"] = df_resp["response_text"].str.lower()

print("Responses:")
display(df_resp.head())




def find_goals_mention(text_lower: str, player_label_lower: str):
    """
    Try to extract a goals number mentioned for a specific player (lowercased).
    Returns an int if found, else None.

    Patterns handled:
      - "player a ... 30 goals"
      - "30 goals ... player a"
      - "player a (30)"
    """
    if not isinstance(text_lower, str):
        return None


    patterns = [

        rf"{re.escape(player_label_lower)}[^0-9]{{0,60}}(\d+)\s+goals?",

        rf"(\d+)\s+goals?[^a-z]{{0,60}}{re.escape(player_label_lower)}",

        rf"{re.escape(player_label_lower)}[^0-9]{{0,10}}\((\d+)\)",
    ]

    for pat in patterns:
        m = re.search(pat, text_lower, flags=re.IGNORECASE | re.DOTALL)
        if m:
            try:
                return int(m.group(1))
            except ValueError:
                continue

    return None


fabrication_records = []

for _, row in df_resp.iterrows():
    resp_lower = row["response_text_lower"]

    for player_label_lower in gt.index.unique():

        g_val = gt.loc[player_label_lower, "G"]
        if isinstance(g_val, pd.Series):

            true_goals = int(g_val.iloc[0])
        else:
            true_goals = int(g_val)


        mentioned_goals = find_goals_mention(resp_lower, player_label_lower)


        if mentioned_goals is not None and mentioned_goals != true_goals:
            fabrication_records.append({
                "prompt_id": row["prompt_id"],
                "hypothesis": row["hypothesis"],
                "condition": row["condition"],
                "model": row["model"],
                "sample_idx": row["sample_idx"],
                "player_id": player_label_lower,
                "true_goals": true_goals,
                "mentioned_goals": mentioned_goals,
            })

fab_df = pd.DataFrame(fabrication_records)
fab_path = "analysis/fabrications_goals.csv"
fab_df.to_csv(fab_path, index=False)
print(f"Saved {fab_path}")
display(fab_df.head())



total_counts = (
    df_resp.groupby(["hypothesis", "condition", "model"])["response_text"]
    .count()
    .reset_index()
    .rename(columns={"response_text": "total_responses"})
)


if not fab_df.empty:
    fab_counts = (
        fab_df.groupby(["hypothesis", "condition", "model"])["player_id"]
        .count()
        .reset_index()
        .rename(columns={"player_id": "fabrication_count"})
    )
else:
    fab_counts = pd.DataFrame(
        columns=["hypothesis", "condition", "model", "fabrication_count"]
    )


merged = pd.merge(
    total_counts,
    fab_counts,
    on=["hypothesis", "condition", "model"],
    how="left",
).fillna({"fabrication_count": 0})

merged["fabrication_rate"] = (
    merged["fabrication_count"] / merged["total_responses"]
)

rates_path = "analysis/fabrication_rates_goals.csv"
merged.to_csv(rates_path, index=False)
print(f"Saved {rates_path}")
display(merged)
```

1. Across all hypotheses, conditions, and models, the fabrication analysis found no instances where a model stated an incorrect number of goals for any player.
2. As a result, the measured fabrication rate was 0.0 for every modelcondition combination in the dataset.
3. This indicates that, under your current prompts and evaluation method, the models consistently avoided numerical hallucinations related to player goal statistics.
